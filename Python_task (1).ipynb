Create a Python program that checks whether a number is prime or not.

Requirements:
- Take input from the user.
- Use a for loop.
- Use if-else statements.
- Print whether the number is prime or not.
- Explain each step with comments.
- Display the expected output.

````python
# Take input from the user and convert it to an integer
num = int(input("Enter a number: "))

# Prime numbers are greater than 1
if num > 1:
    # Iterate from 2 up to (but not including) the number itself
    for i in range(2, num):
        # If the number is divisible by any 'i' in this range, it's not prime
        if (num % i) == 0:
            print(f"{num} is not a prime number.")
            # Break the loop as soon as a divisor is found
            break
    else:
        # If the loop completes without finding any divisors, the number is prime
        print(f"{num} is a prime number.")
# Handle cases where the number is less than or equal to 1
else:
    print(f"{num} is not a prime number.")
````

Create a Python program to print the Fibonacci series.

Requirements:
- Use a while loop.
- Take the number of terms as input.
- Print the Fibonacci sequence.
- Explain every line with comments.
- Display the expected output.

````python
# Take the number of terms for the Fibonacci series as input from the user
num_terms = int(input("Enter the number of terms for the Fibonacci series: "))

# Initialize the first two terms of the Fibonacci series
a, b = 0, 1

# Initialize a counter for the number of terms printed
count = 0

# Check if the number of terms is valid
if num_terms <= 0:
    print("Please enter a positive integer.")
elif num_terms == 1:
    print("Fibonacci sequence up to 1 term:")
    print(a)
else:
    print("Fibonacci sequence:")
    # Use a while loop to generate and print the Fibonacci series
    while count < num_terms:
        print(a, end=" ") # Print the current term, followed by a space
        # Calculate the next term by summing the previous two
        nth = a + b
        # Update the values for the next iteration: 'a' becomes the old 'b', and 'b' becomes the new 'nth'
        a = b
        b = nth
        # Increment the counter
        count += 1
````

Create a Python program for a Shopping Cart System.

Requirements:
- Use dictionaries and lists.
- Allow users to add products.
- Calculate total bill.
- Use for loops and functions.
- Display purchased items.
- Explain every line with comments.
- Show the expected output.

````python
# Define a dictionary to store available products and their prices
products = {
    "apple": 1.00,
    "banana": 0.50,
    "orange": 1.20,
    "milk": 3.00,
    "bread": 2.50
}

# Initialize an empty list to store items in the shopping cart
shopping_cart = []

# Function to display the available products and their prices
def display_products():
    print("\n--- Available Products ---")
    # Iterate through the products dictionary to print each item and its price
    for item, price in products.items():
        print(f"{item.capitalize()}: ${price:.2f}")
    print("------------------------")

# Function to add an item to the shopping cart
def add_to_cart(item_name, quantity):
    # Convert item_name to lowercase to match dictionary keys
    item_name = item_name.lower()
    # Check if the item exists in our products list
    if item_name in products:
        # Add the item and its quantity to the shopping cart list
        shopping_cart.append({"item": item_name, "quantity": quantity, "price": products[item_name]})
        print(f"{quantity} x {item_name.capitalize()} added to cart.")
    else:
        print(f"Sorry, {item_name.capitalize()} is not available.")

# Function to calculate the total bill of items in the shopping cart
def calculate_total_bill():
    total_bill = 0.0
    # Iterate through each item in the shopping cart
    for item_in_cart in shopping_cart:
        # Calculate the cost for each item (quantity * price)
        item_cost = item_in_cart["quantity"] * item_in_cart["price"]
        # Add the item's cost to the total bill
        total_bill += item_cost
    return total_bill

# Function to display the items currently in the shopping cart
def display_cart():
    print("\n--- Your Shopping Cart ---")
    if not shopping_cart:
        print("Your cart is empty.")
        return
    # Iterate through the shopping cart and display each item, quantity, and individual price
    for item_in_cart in shopping_cart:
        print(f"{item_in_cart['item'].capitalize()} x {item_in_cart['quantity']} @ ${item_in_cart['price']:.2f} each")
    print("--------------------------")

# Main program loop for the shopping cart system
print("Welcome to the Shopping Cart System!")
while True:
    print("\nOptions: ")
    print("1. Display Products")
    print("2. Add Item to Cart")
    print("3. View Cart")
    print("4. Checkout")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == '1':
        display_products()
    elif choice == '2':
        item_to_add = input("Enter the name of the item to add: ")
        try:
            qty = int(input(f"Enter quantity for {item_to_add}: "))
            if qty <= 0:
                print("Quantity must be a positive number.")
            else:
                add_to_cart(item_to_add, qty)
        except ValueError:
            print("Invalid quantity. Please enter a number.")
    elif choice == '3':
        display_cart()
    elif choice == '4':
        # Checkout process: display cart, calculate total, and show total bill
        display_cart()
        final_bill = calculate_total_bill()
        print(f"\nYour total bill is: ${final_bill:.2f}")
        print("Thank you for shopping with us!")
        break # Exit the loop after checkout
    elif choice == '5':
        print("Exiting the Shopping Cart System. Goodbye!")
        break # Exit the loop
    else:
        print("Invalid choice. Please enter a number between 1 and 5.")
````

Create a Python Attendance Management System.

Requirements:
- Use dictionaries to store student attendance.
- Use loops to display records.
- Calculate attendance percentage.
- Use functions and conditional statements.
- Add comments explaining each step.
- Display the expected output.

````python
# Initialize an empty dictionary to store student attendance records.
# Each student will be a key, and its value will be another dictionary
# containing 'total_days' and 'present_days'.
attendance_records = {}

# Function to add a new student to the system
def add_student(student_name):
    # Convert the student name to title case for consistent storage
    student_name = student_name.title()
    # Check if the student already exists in the records
    if student_name in attendance_records:
        print(f"Student '{student_name}' already exists.")
    else:
        # If not, add the student with initial attendance counts set to zero
        attendance_records[student_name] = {"total_days": 0, "present_days": 0}
        print(f"Student '{student_name}' added successfully.")

# Function to mark attendance for a specific student
def mark_attendance(student_name, status):
    # Convert the student name to title case for consistent lookup
    student_name = student_name.title()
    # Check if the student exists in the records
    if student_name not in attendance_records:
        print(f"Student '{student_name}' not found. Please add the student first.")
        return

    # Update total days attended for the student
    attendance_records[student_name]["total_days"] += 1
    # If the status is 'P' (Present), increment present_days
    if status.upper() == 'P':
        attendance_records[student_name]["present_days"] += 1
        print(f"Attendance marked 'Present' for '{student_name}'.")
    # If the status is 'A' (Absent), only total_days was incremented
    elif status.upper() == 'A':
        print(f"Attendance marked 'Absent' for '{student_name}'.")
    else:
        # If an invalid status is provided, decrement total_days to revert
        attendance_records[student_name]["total_days"] -= 1 # Revert total_days
        print("Invalid attendance status. Please use 'P' for Present or 'A' for Absent.")

# Function to calculate the attendance percentage for a student
def calculate_attendance_percentage(student_name):
    # Convert the student name to title case for consistent lookup
    student_name = student_name.title()
    # Check if the student exists in the records
    if student_name not in attendance_records:
        return None # Return None if student not found

    # Get attendance data for the student
    data = attendance_records[student_name]
    total = data["total_days"]
    present = data["present_days"]

    # Avoid division by zero if no attendance has been marked yet
    if total == 0:
        return 0.0
    else:
        # Calculate and return the percentage
        return (present / total) * 100

# Function to display all student attendance records
def display_records():
    print("\n--- Attendance Records ---")
    # Check if there are any students in the system
    if not attendance_records:
        print("No student records found.")
        return

    # Iterate through each student in the attendance_records dictionary
    for student, data in attendance_records.items():
        percentage = calculate_attendance_percentage(student)
        # Display student's name, total days, present days, and calculated percentage
        print(f"Student: {student}")
        print(f"  Total Days: {data['total_days']}")
        print(f"  Present Days: {data['present_days']}")
        print(f"  Attendance Percentage: {percentage:.2f}%")
        print("--------------------------")

# Main program loop for the Attendance Management System
print("Welcome to the Attendance Management System!")
while True:
    print("\nOptions:")
    print("1. Add Student")
    print("2. Mark Attendance")
    print("3. View All Records")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        name = input("Enter student name to add: ")
        add_student(name)
    elif choice == '2':
        name = input("Enter student name to mark attendance for: ")
        status = input("Mark attendance (P for Present, A for Absent): ")
        mark_attendance(name, status)
    elif choice == '3':
        display_records()
    elif choice == '4':
        print("Exiting Attendance Management System. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")
````

Create a Python program that implements a simple calculator.

Requirements:
- Use variables and input().
- Perform addition, subtraction, multiplication, and division.
- Use if-elif-else statements.
- Ask the user to choose an operation.
- Display the result.
- Add comments explaining each line.

````python
print("--- Simple Calculator ---")

# Take the first number from the user
num1 = float(input("Enter the first number: "))

# Take the second number from the user
num2 = float(input("Enter the second number: "))

# Ask the user to choose an operation
print("Choose an operation:")
print("1. Addition (+)")
print("2. Subtraction (-)")
print("3. Multiplication (*)")
print("4. Division (/)")
operation = input("Enter choice(1/2/3/4): ")

# Initialize a variable to store the result
result = None

# Use if-elif-else statements to perform the chosen operation
if operation == '1' or operation == '+':
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")
elif operation == '2' or operation == '-':
    result = num1 - num2
    print(f"{num1} - {num2} = {result}")
elif operation == '3' or operation == '*':
    result = num1 * num2
    print(f"{num1} * {num2} = {result}")
elif operation == '4' or operation == '/':
    # Handle division by zero
    if num2 == 0:
        print("Error: Division by zero is not allowed.")
    else:
        result = num1 / num2
        print(f"{num1} / {num2} = {result}")
else:
    print("Invalid input. Please choose a valid operation (1-4 or +, -, *, /).")

print("-------------------------")
````
